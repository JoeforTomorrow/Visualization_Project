```python
lst = []
for i in range(1,100):
    try:
        url = "https://map.naver.com/v5/api/search?"
        payload = {
            "caller": "pcweb",
            "query": "서울 롯데리아",
            "type": "all",
            "searchCoord": "126.8444923;37.5666103",
            "page": i,
            "displayCount": 20,
            "isPlaceRecommendationReplace": "true",
            "lang": "ko"
        }
        response = requests.get(url,params = payload)
        lot_list = response.json()["result"]['place']['list']
        lst = lst + lot_list
    except:
        break
```

```python
lst[0]
```

```python
df = pd.DataFrame()
for lotte in lst:
    name = lotte["name"]
    addr = lotte["address"].split(" ")
    si   = addr[0]
    gu   = addr[1]
    dong = addr[2]
    lat  = lotte["y"]
    lng  = lotte["x"]
    x = pd.Series([name,si,gu,dong,lat,lng])
    df = df.append(x,ignore_index=True)
```

```python
df = df.set_axis(["매장명","시","구","동","위도","경도"],1,inplace=False)
```

```python
df.to_csv("seoul_lotte.csv")
```
