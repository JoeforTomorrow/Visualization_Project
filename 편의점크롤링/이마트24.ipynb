```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
def naver_map_crawler(gu ,search, page):
    url = "https://map.naver.com/v5/api/search?"
    payload = {
        "caller": "pcweb",
        "query": gu + search,
        "type": "all",
        "searchCoord": "",
        "page": page,
        "displayCount": 20,
        "isPlaceRecommendationReplace": "true",
        "lang": "ko",
    }
    headers = {
    'user-agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/101.0.4951.41 Safari/537.36'
    }
    response = requests.get(url, params = payload,headers = headers)
    x = response.json()['result']['place']['list']
    return x
```

```python
lst = []
gus = ['종로구','중구','용산구','성동구',
'광진구','동대문구','중랑구','성북구',
'강북구','도봉구','노원구','은평구',
'서대문구','마포구','양천구','강서구',
'구로구','금천구','영등포구','동작구',
'관악구','서초구','강남구','송파구','강동구']
for gu in gus:
    for i in range(1,100):
        try:
            lst = lst + naver_map_crawler(gu,"이마트24",i)
            print(i, end = " ")
        except:
            break
```

```python
df = pd.DataFrame()
for e24 in lst:
    id   = e24["id"]
    name = e24["name"]
    addr = e24["address"].split(" ")
    si   = addr[0]
    gu   = addr[1]
    dong = addr[2]
    lat  = e24["y"]
    lng  = e24["x"]
    x = pd.Series([id,name,si,gu,dong,lat,lng])
    df = df.append(x,ignore_index=True)
```

```python
df = df.drop_duplicates()
df = df.reset_index()
df = df.drop("index",1)
```

```python
df2 = df.drop("level_0",1)
```

```python
df3 = df2.set_axis(['id','매장명','시','구','동','위도','경도'],1,inplace=False)
```

```python
df_final = df3.drop("id",1)
df_final
```
